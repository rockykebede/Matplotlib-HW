```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

```python
# loading csv files- City and ride csv files
city_data_to_load = "C:/Users/owner/Desktop/GWbootcamp/GWARL201808DATA3/05-Matplotlib/Homework/Instructions/Pyber/data/city_data.csv"
ride_data_to_load = "C:/Users/owner/Desktop/GWbootcamp/GWARL201808DATA3/05-Matplotlib/Homework/Instructions/Pyber/data/ride_data.csv"
```

```python
# Read the City and Ride data and store in dataframes
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
```

```python
#Merging the to dataframe into a single dataframe 
city_ride_combined = pd.merge(ride_data, city_data, on=["city", "city"])
```

```python
# Display the data table for preview - first 5 rows
city_ride_combined.head()
```

```python
#Creating a dataframe Group by city name
groupby_city = city_ride_combined.groupby('city')

groupby_city 
```

```python
# calculating for avg fare by city
avgfare_city = groupby_city.mean()['fare']
avgfare_city.head()
```

```python
# total number of drivers by city
tot_drivers = groupby_city.mean()['driver_count']
tot_drivers.head()
```

```python
#total number of rides(ride_id) by city
totride_bycity = groupby_city['ride_id'].count()
totride_bycity.head()
```

```python
# city by city tpye
city_type = city_data.set_index('city')['type']
city_type.head()
```

```python
# Creating dataframe
City_scatter_final = pd.DataFrame({
    "Total Number of Rides": totride_bycity,
    "Average Fare by City": avgfare_city,
    "Total Number of Drivers": tot_drivers, "City Type": city_type
})
City_scatter_final.head()
```

```python
#colors dictionary
colors = {'Gold':'#FFD700', 'Light Sky Blue':'#87CEFA', 'Light Coral':'#F08080'}
```

```python
#Color Dictionary 
city_color = {'Urban': colors['Light Coral'], 'Suburban': colors['Light Sky Blue'], 'Rural': colors['Gold']}
```

```python
# Scatter plots for each city type
plt.scatter(rural['Total Number of Rides'], rural['Average Fare by City'],
            s = rural['Total Number of Drivers']*10, color = city_color['Rural'],
            edgecolor = 'black', label = 'Rural', alpha = .75)
plt.scatter(suburban['Total Number of Rides'], suburban['Average Fare by City'],
            s = suburban['Total Number of Drivers']*10, color = city_color['Suburban'], 
            edgecolor = 'black', label = 'Suburban', alpha = .75)
plt.scatter(urban['Total Number of Rides'], urban['Average Fare by City'], 
            s = urban['Total Number of Drivers']*10, color = city_color['Urban'], 
            edgecolor = 'black', label = 'Urban', alpha = .75)

#Displaying Bubble plot
## Creating a title, note x label, and y label for our chart

plt.suptitle('Pyber Ride Sharing Ride by City Data (2016)')
plt.title ('Note:Circle Size Correlates with Driver Count per City')
plt.xlabel('Total Number of Rides by City')
plt.ylabel('Total Average Fare by City ($)')

plt.xticks([0,5,10,15,20,25,30,35,40])
plt.yticks([20,25,30,35,40])

# Creating a legend
lgnd = plt.legend(frameon = True, edgecolor = 'black')
lgnd.legendHandles[0]._sizes = [50]
lgnd.legendHandles[1]._sizes = [50]
lgnd.legendHandles[2]._sizes = [50]
plt.savefig("C:/Users/owner/Desktop/Matplotlib_HW/Pyber_ride_share_bubble.png")
plt.show()
```

```python
#group by type to be used by all pie charts
groupby_city = city_ride_combined.groupby('type')['type', 'fare', 'ride_id', 'driver_count' ]

groupby_city.head()
```

```python
#total fare by city
Totfare_bycity = groupby_city.sum()['fare']
Totfare_bycity
```

```python
#pie chart for Total fares by city
labels = Totfare_bycity.index
labels
```

```python
colors = [city_color[n] for n in labels]
explode = [0 , 0, 0.1]
plt.pie(Totfare_bycity, startangle = 180, colors = colors, explode = explode, 
        labels = labels, autopct = "%1.1f%%", shadow = True, wedgeprops = {'linewidth': 0, 'edgecolor': 'black'})

#ploting pie chart
plt.title('% Total Fares by City Type')
plt.axis('equal')
plt.savefig("C:/Users/owner/Desktop/Matplotlib_HW/Total_fare_by_city_tpye.png")
plt.show()
```
